# Importing Required Libraries 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import csv
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
```

# Data Preprocessing

```python
dataset = pd.read_csv("train.csv")    

#dataset from :  https://www.kaggle.com/yassinezayani/flood-prediction/data?select=Train.csv

dataset = dataset.loc[:, dataset.columns != 'Square_ID']
dataset = dataset.loc[:, dataset.columns != 'LC_Type1_mode']

Y = dataset.target_2015
dataset_2 = dataset
dataset = dataset.loc[:, dataset.columns != 'target_2015']
X = dataset
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 12)
X.head()
# X.columns
```

# Data Analysis

```python
for a in ['precip 2014-11-16 - 2014-11-23',
       'precip 2014-11-23 - 2014-11-30', 'precip 2014-11-30 - 2014-12-07',
       'precip 2014-12-07 - 2014-12-14']:
    for b in ['precip 2015-01-04 - 2015-01-11', 'precip 2015-01-11 - 2015-01-18',
           'precip 2015-01-18 - 2015-01-25']:

        sns.regplot(x=a, y=b, data=X_train);
        plt.show()
```

```python
# sns.swarmplot(y=X_train['precip 2015-01-04 - 2015-01-11'],x=y_train)
sns.scatterplot(x=X_train['precip 2015-01-04 - 2015-01-11'], y=X_train['elevation'], hue=y_train)
plt.show()
sns.scatterplot(x=X_train['precip 2015-01-04 - 2015-01-11'], y=X_train['X'], hue=y_train)
plt.show()
sns.scatterplot(x=X_train['precip 2015-01-04 - 2015-01-11'], y=X_train['Y'], hue=y_train)
plt.show()
sns.scatterplot(x=X_train['precip 2015-01-04 - 2015-01-11'], y=X_train['precip 2015-01-11 - 2015-01-18'], hue=y_train)
plt.show()
sns.scatterplot(x=X_train['precip 2015-01-04 - 2015-01-11'], y=X_train['precip 2015-01-18 - 2015-01-25'], hue=y_train)
plt.show()
```

# Training the model

```python
clf = tree.DecisionTreeRegressor(max_depth=16)
clf.fit(X_train,y_train)

print(clf.score(X_test,y_test))
```
